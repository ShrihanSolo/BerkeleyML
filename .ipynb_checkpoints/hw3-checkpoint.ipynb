### 3.1

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal as mnorm
plt.rcParams["figure.figsize"] = (8,8)

def f(x, y, rv):
    return rv.pdf(np.dstack([x, y]))

mn = [1, 1]
cov = [[1, 0], [0, 2]]

x = np.linspace(-2, 4, 1000)
y = np.linspace(-2, 4, 1000)
X, Y = np.meshgrid(x, y)
rv = mnorm(mn, cov)

plt.contourf(X, Y, f(X, Y, rv), 8, alpha=.75, cmap="hot")
C = plt.contour(X, Y, f(X, Y, rv), 8, colors='k')
plt.clabel(C, inline=1, fontsize=10)
plt.gca().set_aspect('equal')
plt.xlabel("x")
plt.ylabel("y")
plt.title("2D Normal Distribution | $\mu = (1, 1)$")


plt.show()
```

## Question 4
Here, $x_1$ and $x_2$ are the random variables.

```python
import numpy as np
np.random.seed(42)
```

```python
x1 = np.random.normal(3, 3, 100)
x2 = 0.5 * x1 + np.random.normal(4, 2, 100)
```

```python
dist = np.vstack([x1, x2])
dist.shape
```

### 4.1

```python
mn = np.mean(dist, axis = 1)
mn
```

### 4.2

Using `numpy.cov`, and $Cov(X) = E[(Z - \mu)(Z - \mu)^T]$, we get two similar results, which likely differ due to rounding errors.

```python
covmat = np.cov(dist, ddof = 0)
covmat
```

```python
cmat_manual = np.matmul((dist.T - mn).T, (dist.T - mn)) / 100
cmat_manual
```

### 4.3

The eigenvalues are calculated below, and the corresponding eigenvectors below that.

```python
eigvals, eigvecs = np.linalg.eig(covmat)
eigvals
```

```python
eigvecs
```

Therefore, the eigenvalues are $9.28952272$ and $2.79607467$, with the corresponding column eigenvectors $[ 0.83732346, 0.54670781]$ and $[ -0.54670781,  0.83732346]$.

### 4.4

```python
import matplotlib.pyplot as plt
```

```python
eigvec_1 = eigvecs.T[0] * eigvals[0]
eigvec_2 = eigvecs.T[1] * eigvals[1]

fig, ax = plt.subplots(1, 1, figsize = [6, 6])
ax.set_title("Eigenvectors of Covariance Matrix")
ax.scatter(dist[0], dist[1], marker = "*", s = 8)
ax.set_ylabel("Random Variable $X_2$")
ax.set_xlabel("Random Variable $X_1$")
ax.set_xlim([-15, 15])
ax.set_ylim([-15, 15])
ax.set_aspect('equal')
origin = np.vstack([mn,mn]).T
a = np.vstack([eigvec_1, eigvec_2])
ax.quiver(*origin, a[:, 0], a[:, 1], color=['r','b'], angles = 'xy', scale_units='xy', scale=1)
plt.show()
```

### 4.5

```python
u = eigvecs
distmn = dist.T - mn
```

```python
dist_new = np.array([np.matmul(u.T, i) for i in distmn]).T
eigvec_1n = np.matmul(u.T, eigvec_1)
eigvec_2n = np.matmul(u.T, eigvec_2)
```

```python
#eigvec_1 = np.array([1, 0]) * eigvals[0]
#eigvec_2 = np.array([0, 1]) * eigvals[1]

fig, ax = plt.subplots(1, 1, figsize = [6, 6])
ax.set_title("Rotated Points with Eigenvector Matrix")
ax.scatter(dist_new[0], dist_new[1], marker = "*", s = 8)
ax.set_ylabel("Random Variable $X_2$")
ax.set_xlabel("Random Variable $X_1$")
ax.set_xlim([-15, 15])
ax.set_ylim([-15, 15])
ax.set_aspect('equal')
origin = np.zeros([2,2])
ax.quiver(*origin, eigvec_1n, eigvec_2n, color=['r','b'], angles='xy', scale_units='xy', scale = 1)
plt.show()
```
